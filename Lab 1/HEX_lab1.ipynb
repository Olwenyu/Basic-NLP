## Question 1: Extracting Data from a PDF File

Using the `PyPDF2` library, write a Python script to extract the entire text from a PDF file. Ensure that you handle cases where the PDF file has multiple pages.

```python
# Installing the PyPDF2 Library
%pip install PyPDF2
```

```python
# import thư viện
import PyPDF2
from PyPDF2 import PdfReader
```

Task Completion
- Find a PDF File with More Than 20,000 Words
- Read the Content and Page Information
- Store the Content in a String Variable

```python
#### YOUR CODE HERE ####

def read_pdf(file_path):
  # Open PDF file
  with open(file_path, 'rb') as pdf_file:
    reader = PdfReader(pdf_file)

    total_pages = len(reader.pages)
    pdf_content = ""

    # Iterate through each page to extract text
    for page_num in range(total_pages):
        page = reader.pages[page_num]
        pdf_content += page.extract_text()

    return pdf_content


file_path = "/content/ED489891.pdf"
try:
  pdf_content = read_pdf(file_path)
  print(pdf_content)
except Exception as e:
  print(f"An error occurred: {e}")

#### END YOUR CODE #####
```

## Question 2: Extracting Data from a DOCX File


```python
# Installing the docx Library
%pip install python-docx
```

```python
#Import library
from docx import Document
```

Task Completion
- Find a DOCX File with More Than 20,000 Words
- Read the Content and Page Information
- Store the Content in a String Variable

```python
#### YOUR CODE HERE ####
def read_docx(file_path):
  # Open DOCX file
  doc = Document(file_path)
  doc_content = ""

  # Iterate through each paragraph to extract text
  for para in doc.paragraphs:
    doc_content += para.text
  return doc_content


file_path = "/content/ED489891.docx"
try:
  doc_content = read_docx(file_path)
  print(doc_content)
except Exception as e:
  print(f"An error occurred: {e}")
#### END YOUR CODE #####
```

### Question 4.2: Complete the String Processing Function

Description of the function: This function takes a string as input and returns a processed version of the string. The main tasks performed in the function are as follows:

- Replace characters matching the pattern `^A-Za-z0-9(),!?\'\`` with a space (" ").
- Replace `\'s` with ` \'s`.
- Replace `\'ve` with ` \'ve`.
- Replace `n\'t` with ` n\'t`.
- Replace `\'re` with ` \'re`.
- Replace `\'d` with ` \'d`.
- Replace `\'ll` with ` \'ll`.
- Replace `,` with ` , `.
- Replace `!` with ` ! `.
- Replace `\(` with ` \( `.
- Replace `\)` with ` \) `.
- Replace `\?` with ` \? `.
- Replace multiple spaces (`\s{2,}`) with a single space.
- Trim leading spaces.
- Convert the text to lowercase.


```python
#### YOUR CODE HERE ####
import re

def process_text(content):
  # Replace characters matching the pattern
  processed = re.sub(r"[^A-Za-z0-9(),!?\'\`]", " ", content)
  processed = re.sub(r"\'s", " \'s", processed)
  processed = re.sub(r"\'ve", " \'ve", processed)
  processed = re.sub(r"n\'t", " n\'t", processed)
  processed = re.sub(r"\'re", " \'re", processed)
  processed = re.sub(r"\'d", " \'d", processed)
  processed = re.sub(r"\'ll", " \'ll", processed)
  processed = re.sub(r",", " , ", processed)
  processed = re.sub(r"!", " ! ", processed)
  processed = re.sub(r"\(", " ( ", processed)
  processed = re.sub(r"\)", " ) ", processed)
  processed = re.sub(r"\?", " ? ", processed)
  processed = re.sub(r"\s{2,}", " ", processed)

  # Trim leading spaces
  processed = processed.strip()

  # Convert the text to lowercase
  processed = processed.lower()

  return processed
#### END YOUR CODE #####
```
